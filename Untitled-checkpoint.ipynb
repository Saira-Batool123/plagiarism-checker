```python
# ================================================================
#                   PLAGIARISM CHECKER TOOL
# ================================================================
# Description : An advanced plagiarism detection tool using
#               TF-IDF Vectorization, Cosine Similarity,
#               Naive Bayes, Decision Tree, and Ensemble Models
#               with GUI, interactive visualizations, export options,
#               and file management.
# Author      : Saira Batool
# Date        : 2025-05-11
# ================================================================

import os
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.ensemble import VotingClassifier
import string
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from tkinter import scrolledtext

# Function to preprocess text
def preprocess_text(text, use_stemming=True, custom_stop_words=None):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    stop_words = set(["the", "and", "is", "in", "on", "for", "to", "of", "a", "an"]) if not custom_stop_words else set(custom_stop_words)
    text = " ".join([word for word in text.split() if word not in stop_words])
    if use_stemming:
        text = " ".join([word[:-1] if word.endswith("ing") else word for word in text.split()])
    return text

# Function to read uploaded text files
def read_uploaded_files(file_names):
    documents = []
    for file in file_names:
        try:
            with open(file, 'r', encoding='utf-8') as f:
                documents.append(f.read())
        except Exception as e:
            messagebox.showerror("Error", f"Failed to read {file}: {e}")
            return None
    return documents if documents else None

# Function to calculate cosine similarity
def calculate_similarity(documents, preprocessor):
    if not documents or len(documents) < 2:
        return None
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([preprocessor(doc) for doc in documents])
    return cosine_similarity(tfidf_matrix)

# Function to prepare training data for classifiers
def prepare_training(similarity_matrix, file_names):
    if similarity_matrix is None or similarity_matrix.shape[0] < 2:
        return None, None, None
    pairs = [(i, j) for i in range(len(file_names)) for j in range(i + 1, len(file_names))]
    labels = [1 if i == j - 1 else 0 for i, j in pairs]  # Simplified labeling
    X, y = [], []
    for (i, j), label in zip(pairs, labels):
        X.append([similarity_matrix[i][j]])
        y.append(label)
    return pairs, X, y

# Function to train models
def train_models(X, y, max_depth=3, min_samples_split=2):
    if X is None or y is None or len(X) < 2:
        return None, None
    try:
        nb = GaussianNB()
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split)
        nb.fit(X, y)
        dt.fit(X, y)
        return nb, dt
    except Exception as e:
        messagebox.showerror("Error", f"Model training failed: {e}")
        return None, None

# Function to create an ensemble model
def create_ensemble(nb, dt, X, y):
    if nb is None or dt is None or X is None or y is None:
        return None
    try:
        ensemble_model = VotingClassifier(estimators=[('nb', nb), ('dt', dt)], voting='hard')
        ensemble_model.fit(X, y)
        return ensemble_model
    except Exception as e:
        messagebox.showerror("Error", f"Ensemble creation failed: {e}")
        return None

# Function to display decision tree
def show_decision_tree(dt, feature_names):
    if dt is not None:
        tree_rules = export_text(dt, feature_names=feature_names)
        return f"\nDecision Tree Representation:\n{tree_rules}"
    return "\nDecision Tree Representation: Not available"

# Function to generate detailed report
def generate_report(similarity_matrix, file_names, pairs, nb_preds, dt_preds, ensemble_preds, dt, threshold):
    report = ["=== Plagiarism Checker Report ==="]
    report.append(f"Date: 2025-05-11")
    report.append(f"Files Analyzed: {len(file_names)}")
    report.append("\nCosine Similarity Matrix:")
    report.append(str(similarity_matrix))
    report.append("\n" + "="*60)
    report.append("Similarity Report:")
    report.append("="*60)
    for idx, (i, j) in enumerate(pairs):
        sim = similarity_matrix[i][j]
        nb_label = "Plagiarized" if nb_preds[idx] else "Not" if nb_preds is not None else "N/A"
        dt_label = "Plagiarized" if dt_preds[idx] else "Not" if dt_preds is not None else "N/A"
        ensemble_label = "Plagiarized" if ensemble_preds[idx] else "Not" if ensemble_preds is not None else "N/A"
        custom_label = "Plagiarized" if sim >= threshold else "Not"
        report.append(f"{os.path.basename(file_names[i])} vs {os.path.basename(file_names[j])} -> Cosine: {sim:.2f} | "
                      f"NB: {nb_label} | DT: {dt_label} | Ensemble: {ensemble_label} | Custom: {custom_label}")
    report.append(show_decision_tree(dt, ['Similarity Score']))
    return "\n".join(report)

# Function to plot heatmap
def plot_heatmap(similarity_matrix, file_names):
    fig, ax = plt.subplots()
    im = ax.imshow(similarity_matrix, cmap='YlOrRd')
    ax.set_xticks(np.arange(len(file_names)))
    ax.set_yticks(np.arange(len(file_names)))
    ax.set_xticklabels([os.path.basename(f) for f in file_names], rotation=45, ha="right")
    ax.set_yticklabels([os.path.basename(f) for f in file_names])
    for i in range(len(file_names)):
        for j in range(len(file_names)):
            ax.text(j, i, f"{similarity_matrix[i, j]:.2f}", ha="center", va="center", color="black")
    plt.colorbar(im)
    ax.set_title("Cosine Similarity Heatmap")
    return fig

# GUI and main logic
def run_plagiarism_checker():
    global file_names, report_content
    root = tk.Tk()
    root.title("Advanced Plagiarism Checker Tool")
    root.geometry("1000x800")

    notebook = ttk.Notebook(root)
    notebook.pack(pady=10, expand=True, fill="both")

    # Upload Tab
    upload_frame = ttk.Frame(notebook)
    notebook.add(upload_frame, text="Upload")
    file_names = []
    uploaded_files = []

    def upload_files():
        files = filedialog.askopenfilenames(filetypes=[("Text files", "*.txt")])
        if files:
            uploaded_files.extend(files)
            file_list.delete(0, tk.END)
            for file in uploaded_files:
                file_list.insert(tk.END, os.path.basename(file))
            preview_button.config(state="normal")
            remove_button.config(state="normal")
            process_button.config(state="normal")
            status_label.config(text=f"Loaded {len(uploaded_files)} files")

    def remove_file():
        selected = file_list.curselection()
        if selected:
            idx = selected[0]
            uploaded_files.pop(idx)
            file_list.delete(idx)
            preview_text.delete(1.0, tk.END)
            status_label.config(text=f"Loaded {len(uploaded_files)} files")
            if not uploaded_files:
                preview_button.config(state="disabled")
                remove_button.config(state="disabled")
                process_button.config(state="disabled")

    def preview_file():
        selected = file_list.curselection()
        if selected:
            idx = selected[0]
            with open(uploaded_files[idx], 'r', encoding='utf-8') as f:
                preview_text.delete(1.0, tk.END)
                preview_text.insert(tk.END, f.read())

    tk.Button(upload_frame, text="Upload Text Files", command=upload_files).pack(pady=5)
    file_list = tk.Listbox(upload_frame, height=10, width=50)
    file_list.pack(pady=5)
    preview_button = tk.Button(upload_frame, text="Preview Selected File", command=preview_file, state="disabled")
    preview_button.pack(pady=5)
    remove_button = tk.Button(upload_frame, text="Remove Selected File", command=remove_file, state="disabled")
    remove_button.pack(pady=5)
    preview_text = scrolledtext.ScrolledText(upload_frame, height=10, width=50)
    preview_text.pack(pady=5)

    # Results Tab with Scrollable Frame
    results_frame = ttk.Frame(notebook)
    notebook.add(results_frame, text="Results")

    # Create a canvas and scrollbar for the Results tab
    canvas = tk.Canvas(results_frame)
    scrollbar = ttk.Scrollbar(results_frame, orient="vertical", command=canvas.yview)
    scrollable_frame = ttk.Frame(canvas)

    # Configure the canvas
    scrollable_frame.bind(
        "<Configure>",
        lambda e: canvas.configure(scrollregion=canvas.bbox("all"))
    )
    canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")
    canvas.configure(yscrollcommand=scrollbar.set)

    # Pack the canvas and scrollbar
    canvas.pack(side="left", fill="both", expand=True)
    scrollbar.pack(side="right", fill="y")

    # Add widgets to the scrollable frame
    result_text = scrolledtext.ScrolledText(scrollable_frame, height=20, width=100)
    result_text.pack(pady=10)
    canvas_frame = ttk.Frame(scrollable_frame)
    canvas_frame.pack(pady=10)
    export_button = tk.Button(scrollable_frame, text="Export Report", state="disabled")
    export_button.pack(pady=5)

    # Settings Tab
    settings_frame = ttk.Frame(notebook)
    notebook.add(settings_frame, text="Settings")
    tk.Label(settings_frame, text="Similarity Threshold:").pack()
    threshold_scale = tk.Scale(settings_frame, from_=0.1, to=1.0, resolution=0.1, orient=tk.HORIZONTAL, length=200)
    threshold_scale.set(0.5)
    threshold_scale.pack(pady=5)
    tk.Label(settings_frame, text="Max Tree Depth:").pack()
    max_depth_scale = tk.Scale(settings_frame, from_=3, to=10, resolution=1, orient=tk.HORIZONTAL, length=200)
    max_depth_scale.set(3)
    max_depth_scale.pack(pady=5)
    tk.Label(settings_frame, text="Min Samples Split:").pack()
    min_samples_scale = tk.Scale(settings_frame, from_=2, to=10, resolution=1, orient=tk.HORIZONTAL, length=200)
    min_samples_scale.set(2)
    min_samples_scale.pack(pady=5)
    tk.Checkbutton(settings_frame, text="Enable Stemming", variable=tk.BooleanVar(value=True)).pack(pady=5)
    tk.Label(settings_frame, text="Custom Stop Words (comma-separated):").pack()
    stop_words_entry = tk.Entry(settings_frame)
    stop_words_entry.pack(pady=5)

    # Help Tab
    help_frame = ttk.Frame(notebook)
    notebook.add(help_frame, text="Help")
    tk.Label(help_frame, text="Instructions:\n1. Upload .txt files in the Upload tab.\n2. Preview or remove files if needed.\n3. Adjust settings in the Settings tab.\n4. Process in the Results tab to see analysis.\n5. Scroll down in the Results tab to export the report to your PC.").pack(pady=10)
    tk.Label(help_frame, text="About:\nPlagiarism Checker Tool by Roll Numbers 507, 542, 573, 596\nPresented to: Mam Habiba\nDate: 2025-05-11").pack(pady=10)

    # Progress Bar
    progress = ttk.Progressbar(root, length=400, mode='determinate')
    progress.pack(pady=10)

    # Process button and logic
    def process():
        global report_content
        if not uploaded_files:
            messagebox.showerror("Error", "Please upload at least two files.")
            return
        progress['value'] = 0
        root.update()
        custom_stop_words = stop_words_entry.get().split(",") if stop_words_entry.get() else None
        preprocessor = lambda x: preprocess_text(x, use_stemming=tk.BooleanVar(value=True).get(), custom_stop_words=custom_stop_words)
        documents = read_uploaded_files(uploaded_files)
        if documents is None:
            return
        progress['value'] = 20
        root.update()
        similarity_matrix = calculate_similarity(documents, preprocessor)
        if similarity_matrix is None:
            messagebox.showerror("Error", "Insufficient documents for analysis.")
            return
        progress['value'] = 40
        root.update()
        print("\nCosine Similarity Matrix:")
        print(similarity_matrix)
        pairs, X, y = prepare_training(similarity_matrix, uploaded_files)
        if X is None or y is None or len(X) < 2:
            messagebox.showerror("Error", "Insufficient data for training. Need at least 2 pairs.")
            return
        progress['value'] = 60
        root.update()
        nb, dt = train_models(X, y, max_depth=max_depth_scale.get(), min_samples_split=min_samples_scale.get())
        if nb is None or dt is None:
            # Fallback to cosine similarity alone
            nb_preds = dt_preds = ensemble_preds = [None] * len(pairs)
        else:
            ensemble_model = create_ensemble(nb, dt, X, y)
            nb_preds = nb.predict(X) if nb else [None] * len(pairs)
            dt_preds = dt.predict(X) if dt else [None] * len(pairs)
            ensemble_preds = ensemble_model.predict(X) if ensemble_model else [None] * len(pairs)
        progress['value'] = 80
        root.update()
        progress['value'] = 100
        root.update()

        # Generate and display results
        report_content = generate_report(similarity_matrix, uploaded_files, pairs, nb_preds, dt_preds, ensemble_preds, dt, threshold_scale.get())
        result_text.delete(1.0, tk.END)
        result_text.insert(tk.END, report_content)

        # Plot heatmap
        fig = plot_heatmap(similarity_matrix, uploaded_files)
        for widget in canvas_frame.winfo_children():
            widget.destroy()
        canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
        canvas.draw()
        canvas.get_tk_widget().pack()

        # Enable export button
        export_button.config(state="normal")
        status_label.config(text="Processing complete!")
        progress['value'] = 0

    def export_report():
        global report_content
        if not report_content:
            messagebox.showerror("Error", "No report available to export. Please process files first.")
            return
        save_path = filedialog.asksaveasfilename(defaultextension=".txt", filetypes=[("Text files", "*.txt")])
        if save_path:
            try:
                with open(save_path, 'w', encoding='utf-8') as f:
                    f.write(report_content)
                messagebox.showinfo("Success", f"Report saved to {save_path}")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to save report: {e}")

    export_button.config(command=export_report)
    process_button = tk.Button(root, text="Process", command=process, state="disabled")
    process_button.pack(pady=10)
    status_label = tk.Label(root, text="No files loaded")
    status_label.pack()

    report_content = None
    root.mainloop()

if __name__ == "__main__":
    run_plagiarism_checker()
```
